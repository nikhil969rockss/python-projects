# Load the book

```python
with open("miracle_in_the_andes.txt", "r") as file:
    book = file.read()
```

```python
import nltk
nltk.download('vader_lexicon')
from nltk.sentiment import SentimentIntensityAnalyzer
analyser = SentimentIntensityAnalyzer()
```

### example 

```python
analyser.polarity_scores("hey your are beautiful and i love your eyes")
```

# analyzie the whole book

```python
analyser.polarity_scores(book)
```

- The book has both equally positive and negative chapters to balanced book

# analyse the individual chapters

```python
import re

pattern = re.compile("Chapter [0-9]+")
chapters = re.split(pattern, book)
chapters = chapters[1:]

for number,chapter in enumerate(chapters):
    scores = analyser.polarity_scores(chapter)
    print("Chapter", number+1, scores)
```
