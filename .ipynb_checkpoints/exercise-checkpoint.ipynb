# Load the Book

```python
with open("miracle_in_the_andes.txt", "r") as file:
    book = file.read()
```

# Extract the paragraph where the 'love' was used

```python
import re
pattern = re.compile("[^\n]+love[^\n]+")
paragraphs = re.findall(pattern, book)
paragraphs
len(paragraphs)
```

# Extract the title of the book

### Mehod 1

```python
pattern = re.compile("[a-zA-Z ,]+\n\n")
findings = re.findall(pattern, book)
findings = [i.strip("\n") for i in findings]
findings
```

### method-2

```python
pattern = re.compile("([a-zA-Z ,]+)\n\n")
findings = re.findall(pattern, book)
findings
```

# Funtion that finds out occurrance of any word

```python
def find(w):
    pattern = re.compile("[a-zA-Z]+")
    findings = re.findall(pattern, book.lower())

    d={}
    for word in findings:
        if word in d.keys():
            d[word] = d[word] + 1
        else:
            d[word] = 1

    try:
        return d[w]
    except:
        return f"{w} does not exist in the book"
        
    
```

# call the function

```python
find("love")
```
