# Load the book

```python
with open("miracle_in_the_andes.txt", "r") as file:
    book = file.read()
```

# Find the most used word (non-articles)

```python
import re

pattern = re.compile("[a-zA-Z]+")
findings = re.findall(pattern, book.lower())
findings[:10]
```

### count the words

```python
d = {}

for word in findings:
    if word in d.keys():
        d[word] = d[word] + 1
    else:
        d[word] = 1
```

```python
word_list = [(values,keys) for (keys, values) in d.items()]
word_list = sorted(word_list, reverse=True)
word_list[:10]
```

```python
import nltk
nltk.download("stopwords")
from nltk.corpus import stopwords
english_stopwords = stopwords.words("english")

```

### excluding these words from the above list

```python
new_list =[]
for count,word in word_list:
    if word not in english_stopwords:
        new_list.append((word, count))

new_list[:10]
```
